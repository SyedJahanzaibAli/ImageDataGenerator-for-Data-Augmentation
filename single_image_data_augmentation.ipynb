```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/human.jpg', target_size=(200, 200))
```

```python
img
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip= True)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape((1,200,200,3))
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/content', save_prefix='human', save_format='jpg'):
    i += 1
    if i > 20:
        break
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Directory where your images are stored
content_dir = '/content/'

# Get list of all image files in the directory
image_files = [f for f in os.listdir(content_dir) if os.path.isfile(os.path.join(content_dir, f))]

# Ensure we only take the first 16 images if there are more than 16
num_images = min(len(image_files), 16)

# Create a figure to contain the plot
fig = plt.figure(figsize=(10, 10))

# Loop through the images and plot them in a 4x4 grid
for i in range(num_images):
    img_path = os.path.join(content_dir, image_files[i])
    img = mpimg.imread(img_path)
    fig.add_subplot(4, 4, i + 1)  # subplot index starts from 1
    plt.imshow(img)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
